## groupby().aggfun(), .xs(key=,level=), .swaplevel(), sort_index(level= ) , .agg()

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('mpg.csv')
df
```

```python
df.groupby('model_year').mean()[['mpg','displacement']]
```

```python
year_cyl = df.groupby(['model_year','cylinders']).mean()
```

```python
year_cyl.index.names
```

```python
year_cyl.index.levels
```

```python
year_cyl.loc[[70,82]]
```

```python
year_cyl.loc[(70,6)]
```

```python
year_cyl.xs(key=70,level='model_year')
```

```python
year_cyl.xs(key=4,level='cylinders')
```

```python
year_cyl.swaplevel()
```

```python
year_cyl.sort_index(level='model_year',ascending=True)
```
