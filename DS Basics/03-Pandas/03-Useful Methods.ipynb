## Apply function

```python
def last_four(num):
    return str(num)[-4:]
    
```

```python
last_four(5476798798)
```

```python
df['CC last_four'] = df['CC Number'].apply(last_four)
```

```python
df
```

```python
df['total_bill'].mean()
```

```python
def yelp(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 30:
        return '$$'
    else:
        return '$$$'
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
```

### Applying on multiple columns

```python
def quality(total_bill,tip):
    if tip/total_bill > 0.25:
        return 'Generous'
    else:
        return 'Other'
```

```python
quality(16.99,1.01)
```

```python
# df['quality'] = df[['total_bill','tip']].apply(lambda df:quality(df['total_bill'],df['tip']),axis=1)
```

#### Applying np.vectorize method

```python
df['quality'] = np.vectorize(quality)(df['total_bill'],df['tip'])
```
